```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
my_data = pd.read_csv('E:\Data Science\Python\machine learning exercises\dataset\house_price.csv')
my_data
```

### Membuat filter correlation features manual

```python
corr_matrix = my_data.drop(labels='medv', axis=1).corr()
corr_matrix
```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(corr_matrix, annot=True,fmt='.2f')
```

```python
14 * 14
```

```python
treshold = 0.4
correlated_feature = set()

for i in range(len(corr_matrix.columns)):
    for j in range(len(corr_matrix.columns)):
        if i == j:
            continue
        else:
            if abs(corr_matrix.iloc[i, j]) > treshold:
                correlated_feature.add(corr_matrix.columns[i])
                correlated_feature.add(corr_matrix.columns[j])
```

```python
correlated_feature
```

### Membandingkan hasil filter correlation features dengan package jcopml

```python
from jcopml.plot import plot_correlation_matrix, plot_actual_vs_prediction
```

```python
plot_correlation_matrix(df=my_data, target_col='medv')
```

### Modelling

```python
from sklearn.linear_model import LinearRegression
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.pipeline import Pipeline
from jcopml.pipeline import num_pipe, cat_pipe
from jcopml.tuning import grid_search_params as gsp
from jcopml.plot import plot_missing_value, plot_residual
from jcopml.feature_importance import mean_score_decrease
```

```python
plot_missing_value(my_data)
```

```python
gsp.linreg_params
```

```python
X = my_data[['indus', 'rm', 'ptratio', 'lstat']]
y = my_data.medv

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
preprocessor = ColumnTransformer([
    ('numeric', num_pipe(poly=2, scaling='standard'), list(X_train.columns))
])
    
pipeline = Pipeline([
    ('prep', preprocessor),
    ('algo', LinearRegression())
])

model = GridSearchCV(estimator=pipeline, param_grid=gsp.linreg_params, cv=3, n_jobs=-1, verbose=1)
model.fit(X_train, y_train)

print(model.best_params_)
print(model.best_score_, model.score(X_train, y_train), model.score(X_test, y_test))
```

```python
importance = mean_score_decrease(X_train, y_train, model, plot=True)
```

```python
plot_actual_vs_prediction(X_train, y_train, X_test, y_test, model)
```

```python
plot_residual(X_train, y_train, X_test, y_test, model)
```
