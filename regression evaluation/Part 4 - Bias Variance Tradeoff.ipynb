### Random Forest

1. n_estimator -> more averaging -> decrease overfit
2. increase max_depth -> prone to overfit
3. increase min_samples_split -> prevent splitting -> less depth -> decrease overfit
4. increase max_feature -> prone to get best decision -> prone to overfit
5. increase min_impurity (penurunan error sebelum dilakukan splitting) -> prevent splitting -> decrease overfit
6. increase min_samples_leaf -> decrease overfit
